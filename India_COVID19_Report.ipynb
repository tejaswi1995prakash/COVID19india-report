```
statewise = pd.read_csv('https://api.covid19india.org/csv/latest/state_wise_daily.csv')
```

```
statewise_test = pd.read_csv('https://api.covid19india.org/csv/latest/statewise_tested_numbers_data.csv')
```

```
statewise['Date'] = pd.to_datetime(statewise['Date'])
```

```
states_only = statewise.drop(columns='TT')
```

```
s1 = pd.melt(states_only,id_vars=['Date','Status']).sort_values(['Date','Status','variable'])
s1.head()
```

```
ss = pd.crosstab(s1['variable'],s1['Status'],values= s1['value'],aggfunc= np.sum)
ss.reset_index()
ss["Mortality_Rate"]=(ss['Deceased']/ss['Confirmed'])
ss.sort_values('Mortality_Rate',ascending=False)
```

We can now see the total number of Confirmed,Deceased and Recovered cases state wise.
Though Meghalaya has the highest Mortality rate, It should be noted that the sample size is very less in Meghalaya. Madhya Pradesh on the other hand has a mortality rate around 5.5% which is followed by West Bengal which is around 5.2%

# **Visualization**
# Statewise representation 

```

import plotly.graph_objects as go

fig = go.Figure([go.Bar(x=s1['variable'].unique(), y=ss['Confirmed'])])
fig.show()
```

```
import plotly.graph_objects as go
#states = ([s1['variable'].unique()])


fig = go.Figure()
fig.add_trace(go.Bar(x=s1['variable'].unique(),
                y=ss['Confirmed'],
                name='Total Cases',
                marker_color='rgb(15, 83, 109)'
                ))
fig.add_trace(go.Bar(x=s1['variable'].unique(),
                y=ss['Deceased'],
                name='Deaths',
                marker_color='rgb(26, 118, 255)'
                ))
fig.update_layout(
    title='COVID19 INDIA',
    xaxis_tickfont_size=14,
    yaxis=dict(
        title='Number of cases',
        titlefont_size=16,
        tickfont_size=14,
        ),

    legend=dict(
        x=0,
        y=1.0,
        bgcolor='rgba(250, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    ),

  barmode='group',
    bargap=0.15, # gap between bars of adjacent location coordinates.
    bargroupgap=0.1 # gap between bars of the same location coordinate.
)
fig.show()
```
